```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

base = Path(r"C:\Python Projects\Financial Analysis")
df = pd.read_csv(base / "Finance_data.csv")

sns.set_style("darkgrid")
print(df.head())
```

```python
df.head()
```

```python
df.drop(['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'PPF', 'Fixed_Deposits', 'Gold'], axis = 1, inplace = True)
```

```python
df = df.replace(r'^\s*$', np.nan, regex = True)
```

```python
sns.countplot(
    x='gender',
    data=df,
    hue='gender',        # add this
    palette='Set2',
    linewidth=3,
    edgecolor='black',
    legend=False         # prevents duplicate legend
)
plt.show()
```

```python
sns.countplot(x='age', data=df, palette='Set3', hue='gender', linewidth=2, edgecolor='black')
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.countplot(x=df['gender'], hue = df['Investment_Avenues'], palette = 'summer', linewidth = 3, edgecolor = 'white')
plt.title('INVESTMENT AVENUE')

plt.subplot(1,2,2)
sns.countplot(x= df['gender'], hue = df['Stock_Marktet'], palette = 'hot', linewidth = 3, edgecolor = 'white')
plt.title('STOCK MARKET')
plt.show()
```

```python
sns.countplot(
    x='Factor',
    data=df,
    hue='Factor',
    palette='coolwarm',
    linewidth=2,
    edgecolor='black',
    legend=False
)
plt.show()
```

```python
plt.figure(figsize = (16,8))
sns.pointplot(x = 'Purpose', y = 'age', data = df, linestyles = '--', capsize  = .3, color = 'Blue')
plt.show()
```

```python
plt.figure(figsize = (18,6))
plt.subplot(1,2,1)
sns.countplot(hue = df ['gender'], x = df['Duration'], palette = 'viridis', linewidth = 2, edgecolor = 'black')
plt.title ('DURATION')

plt.subplot(1,2,2)
sns.countplot(hue = df['gender'], x = df['Invest_Monitor'], palette ='seismic', linewidth = 2, edgecolor = 'black')
plt.title('INVESTMENT MONITORING')
plt.show()
```
